**Question 1**

Write a program that calculates the amount of money a person would earn over a period of time if his or her salary is one penny the first day, two pennies the second day, and continues to double each day. The program should ask the user for the number of days. Display a table showing what the salary was for each day, then show the total pay at the end of the period. The output should be displayed in a dollar amount, not the number of pennies.

**Grading Rubric**

* Ask user for input - 1point
* Double pennies in dollars - 2points
* Calculate total in dollars - 2points
* Table - 2points
* Sufficient comments - 2 points
* Correct output - 1points


```python
# ask user number of days
days = int(input("Enter the number of days to: "))

# print headings
print()
print("Days\t\tSalary in $")
print("--------------------------")

# set an accumulator and create loop structure
# the first day is one penny
salary = 0.01 
total=0
for d in range(1,days+1,1):
    print(d,"\t\t$", salary)
    total += salary
    salary *= 2
    
# show total   
print("--------------------------")
print(f"Total Salary:\t${total:.2f}")
```

**Question 2**

A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself. The first few prime numbers are 2, 3, 5, 7, 11, . . . .. 

Given two positive integers start and end. The task is to write a Python program to print all Prime numbers in an Interval.


**Grading Rubric**

* Ask user for input - 2point
* Correct use of for-loops - 2points
* Correct use of remainder division - 2points
* Sufficient comments - 2 points
* Correct output - 1points


```python
start = int(input("Enter start number: "))
end = int(input("Enter end number: "))

for num in range(start, end + 1):
  # If given number is greater than 1
  if num > 1:
      # Iterate from 2 to num / 2
      for i in range(2, int(num/2) + 1):
          # If num is divisible by any number between
          # 2 and num / 2, it is not prime
          if (num % i) == 0:
              break
      else:
          print(num)
  
```

**Question 3**

Create a program that asks the user to enter in a number and tells us if our number is even or odd. If the number is even, check to see if it is also divisible by 12 and if the number is odd check to see if it is divisible by 5.

**Grading Rubric**

* Ask user for input - 2points
* Correct use of if-then stateme - 2points
* Correct use of remainder division - 2points
* Sufficient comments - 2 points
* Correct output - 1points


```python
num = int(input("Enter number: "))

if num % 2 == 0:
  if num % 12 == 0:
    print("The number is even and is also divisible by 12")
  else:
    print("The number is even but not divisible by 12")
else:
  if num % 3 == 0:
    print("The number is odd and is also divisible by 3")
  else:
    print("The number is odd but not divisible by 3")
```
