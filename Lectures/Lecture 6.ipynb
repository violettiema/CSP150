```python
# named constants
base_hours = 100
overtime_multiplier = 1.5
pay_rate = float(input("What is your pay rate? "))
hours = float(input("How many hours did you work this week? "))

# Calculate and display the check/gross pay
if hours > base_hours:
  #calculate overtime hours
  overtime_hours = hours - base_hours

  # calculate the overtime pay
  overtime_pay = overtime_hours * pay_rate * overtime_multiplier

  # calculate the gross pay
  gross_pay = (base_hours * pay_rate) + overtime_pay

else:
  gross_pay = hours * pay_rate
# display the paycheck
print(f'Your gross pay is ${gross_pay}')
```
