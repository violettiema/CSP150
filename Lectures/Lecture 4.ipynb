General format

{placeholdername:format - specifier}

```python
# rounding off floating point numbers
# 2f means we are rounding off to 2 decimal places
# return a floating point number
num = 123.456789
print(f'{num:.2f}')
```

```python
a = 2
b = 3
print(a/b)
```

```python
print(f'{a/b:.4f}')
```

```python
# insert comma separators
numb = 123456789.123456789
print(f'{numb:,}')
```

```python
# insert dollar signs/symbols
monthly_pay = 12000.0
annual_pay = monthly_pay * 12
print(f'This is annual salary ${annual_pay:.2f}')
```

```python
print('This is annual salary $',annual_pay)
```

```python
print(f'This is annual salary ${annual_pay:,.2f}')
```

```python
# format a float as a percentage
discount = 0.555
print(f'The percentage discount is {discount:.06%}')
```

```python
# formatting to scientific notation
num = 123456.123456
print(f'{num:e}')
```

```python
# formating integers - we use d
number = 1234
print(f'{number:d}')
```

```python
print(f'{number:,d}')
```

```python
# This program displays the numbers in two columns
num1 = 123.456
num2 = 79.24363982
num3 = 4567.1
num4 = 1.23
num5 = 1234.5678
num6 = 15.12345
# we want to display each number in a field of 10spaces
# we want to round off to 2 decimal places
print(f'{num1:10.2f}{num2:10.2f}')
print(f'{num3:10.2f}{num4:10.2f}')
print(f'{num5:10.2f}{num6:10.2f}')
```

Aligning values
* \< = left aligns 
* \> = right aligns
* \^ = center align

```python
# This program demonstrates the use of the center align
name1 = "Blake"
name2 = "Noah"
name3 = "Chelsea"
name4 = "Jeeva"
name5 = "Josh"
name6 = "Jermaine"
name7 = "Richard"
# display the names
print(f'****{name1:^20}****')
print(f'****{name2:^20}****')
print(f'****{name3:^20}****')
print(f'****{name4:^20}****')
print(f'****{name5:^20}****')
print(f'****{name6:^20}****')
print(f'****{name7:^20}****')
```

```python
# The order of designators
# [alignment][width][,][.precision][type]
example = 123456789.123456789
print(f'We are trying to see this one {example:>30,.3f}')
```

```python
print(f'We are trying to see this one {example:30^,.3f}')
```
