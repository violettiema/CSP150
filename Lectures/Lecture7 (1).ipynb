**Compare strings**

Python allows you to compare strings. This allows you to create a decision structure that can test the value of that string.

```python
# This program is to compare if two strings are equal
name1 = "Emmanuel"
name2 = "Jose"
if name1 == name2:
  print("The names are the same")
else:
  print("The names are not the same")
```

```python
# This program is to compare if two strings are not equal
month = "September"
if month != "October":
  print("This is the wrong month for Octoberfest!")
else:
  print("Huraaaay! Lets PAARRRTTTYYYY")
```

**Nested Decision structures and if-elif-else statement**

* To test more than one condition, a decision structure must be used. 

```python
# This program determines if a bank customer 
# qualifies for a loan

min_salary = 10000.0    # minimum annual salary
min_years = 2            # minimum #years at the job

# Get the customers annual salary
salary = float(input("Enter your annual salary: "))

# Get the number of years on current job
years = float(input("Enter the number of years at your current job: "))

# Determine whether the customer qualifies
if salary >= min_salary:
  if years >= min_years:
    print("You qualify for the loan!!!!!!")
  else:
    print("You must have",min_years,"years to qualify")
else:
  print(f"You must have a minimum of ${min_salary:,.2f} to qualify")
```
