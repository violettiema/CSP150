Aligning Values
* \< = left aligns the value
* \> = right aligns the value
* \^ = center aligns the value

```python
# This program demonstrates how to center align strings
name1 = "Danielle"
name2 = "Spencer"
name3 = "Rohan"
name4 = "Emmanuel"
name5 = "Jose"
name6 = "Joshua"

# Display the names
print(f'****{name1:^20}****')
print(f'****{name2:^20}****')
print(f'****{name3:^20}****')
print(f'****{name4:^20}****')
print(f'****{name5:^20}****')
print(f'****{name6:^20}****')
```

```python
# The order of designators
# [alignment][width][,][.precision][type]
example = 123456.789123
print(f'We are trying to see this {example:^30,.2f}')
```

```python
# This example is to show what happens when you misuse the order
print(f'We are trying to see this {example:30<,.2f}')
```
