```python
# named constants 
base_hours = 100
overtime_multiplier = 1.5

# Get number of hours worked
hours = float(input("How many hours did you work?"))
pay_rate = float(input("What is your pay rate?"))

# Calculate and display the check
if hours > base_hours:
  #calcuate the difference hours for overtime
  overtime_hours = hours - base_hours

  # calculate the overtime pay
  Overtime_pay = overtime_hours * overtime_multiplier * pay_rate

  # Calculate gross pay
  gross_pay = Overtime_pay + base_hours * pay_rate

else:
  gross_pay = base_hours * pay_rate

# Display the gross pay
print(f'The amount you earned is ${gross_pay:,.2f}')
```
